## 1.1 Contexto


A AeroVision, empresa especializada em soluções analíticas para o setor aéreo, está conduzindo uma pesquisa de mercado com o objetivo de identificar aeroportos que mais se beneficiariam de soluções baseadas em dados.

A partir dessa análise, foram respondidas perguntas estratégicas que ajudam a AeroVision a:

Identificar aeroportos com alta complexidade e volume operacional;

Detectar tendências que indiquem necessidade de modernização;

Avaliar onde soluções analíticas trariam maior impacto operacional e financeiro.

## 1.2 Objetivos do Projeto

Com base na análise de dados sobre passageiros e voos, o projeto busca levantar insights de impacto direto na qualidade do serviço, eficiência de operação e fidelização de clientes, respondendo às seguintes perguntas de negócio:

**Quais perfis demográficos de passageiros estão mais associados a voos com atraso ou cancelamento?**



**Quais rotas, países ou continentes concentram a maioria dos atrasos?**



**Quais pilotos ou grupos operacionais estão mais associados a voos com problemas?**



**O padrão de atrasos varia por horário ou data específica?**

## 1.3 Benefícios Esperados


**Aprimoramento da experiência do passageiro**, com foco em públicos estratégicos.

**Maior eficiência operacional**, com foco em decisões baseadas em dados.

**Forças (Strengths)**

Riscos operacionais externos: Condições climáticas, problemas logísticos ou crises geopolíticas podem impactar a regularidade dos voos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir estilo mais moderno
plt.style.use('ggplot')

# Extrair a contagem de passageiros por Nacionalidade
nationality_counts = df['Nationality'].value_counts()

# Criar um novo DataFrame com as nacionalidades e contagens
df_nationalities = pd.DataFrame({
    'Nacionalidade': nationality_counts.index,
    'Número de Passageiros': nationality_counts.values
})

# Ordenar o DataFrame pela contagem em ordem decrescente
df_nationalities = df_nationalities.sort_values(by='Número de Passageiros', ascending=False)

# Criar figura com tamanho adequado
plt.figure(figsize=(18, 10))

# GRÁFICO 1: Top 15 nacionalidades (ao invés de todas que ficariam ilegíveis)
plt.subplot(1, 2, 1)
top_15 = df_nationalities.head(15)

# Usar uma nova paleta de cores vibrantes
bars = plt.barh(top_15['Nacionalidade'], top_15['Número de Passageiros'], 
                color=sns.color_palette("husl", 15))  # Paleta "husl" para cores vibrantes

# Adicionar valores nas barras
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2, 
             f'{top_15["Número de Passageiros"].iloc[i]:,}',
             va='center', fontsize=9)

# Configurar os rótulos e título
plt.title('Top 15 Nacionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Número de Passageiros', fontsize=12)
plt.gca().invert_yaxis()  # Inverter o eixo y para mostrar a maior contagem no topo
plt.grid(axis='x', linestyle='--', alpha=0.7)

# GRÁFICO 2: Distribuição por grupos de nacionalidades
plt.subplot(1, 2, 2)

# Criar grupos mais significativos
top_10 = df_nationalities.head(10)
next_10 = df_nationalities.iloc[10:20]
other_nationalities = df_nationalities.iloc[20:]

# Criar dados agregados para o gráfico de pizza
sizes = [
    top_10['Número de Passageiros'].sum(),
    next_10['Número de Passageiros'].sum(),
    other_nationalities['Número de Passageiros'].sum()
]

labels = [
    f'Top 10 Nacionalidades\n{sizes[0]:,} passageiros ({sizes[0]/sum(sizes):.1%})',
    f'Nacionalidades 11-20\n{sizes[1]:,} passageiros ({sizes[1]/sum(sizes):.1%})',
    f'Outras Nacionalidades\n{sizes[2]:,} passageiros ({sizes[2]/sum(sizes):.1%})'
]

# Criar gráfico de pizza com novas cores
explode = (0.05, 0.05, 0.05)
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', 
                                   startangle=90, explode=explode, 
                                   colors=['orange', 'cyan', 'purple'],  # Novas cores vibrantes
                                   shadow=True, textprops={'fontsize': 11})

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.axis('equal')
plt.title('Distribuição por Grupo de Nacionalidades', fontsize=14, fontweight='bold')

# Adicionar detalhes das top 5 nacionalidades em uma caixa de texto
props = dict(boxstyle='round', facecolor='lightyellow', alpha=0.5)  # Cor de fundo diferente
top_5_info = "\n".join([f"{row['Nacionalidade']}: {row['Número de Passageiros']:,}" 
                        for _, row in top_15.head(5).iterrows()])
plt.text(1.4, 0, 'Top 5 Nacionalidades:\n\n' + top_5_info, fontsize=10,
         verticalalignment='top', horizontalalignment='center',
         transform=plt.gca().transAxes, bbox=props)

# Ajustes finais
plt.tight_layout()
plt.suptitle('Análise por Nacionalidade', fontsize=16, fontweight='bold', y=1.02)

# Adicionar informações úteis no rodapé
plt.figtext(0.5, -0.01, 'Total de nacionalidades analisadas: ' + str(len(df_nationalities)) + 
            ' | Total de passageiros: ' + f'{df_nationalities["Número de Passageiros"].sum():,}',
            ha='center', fontsize=10)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que você já tenha o DataFrame 'df' carregado
# df = pd.read_csv('seu_arquivo.csv')  # Exemplo de como carregar o DataFrame

# Configurar estilo do Seaborn
sns.set(style="whitegrid")

# Criar figura com tamanho adequado
plt.figure(figsize=(14, 6))

# GRÁFICO 1: Histograma
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title('Histograma', fontsize=14)
plt.xlabel('Idade', fontsize=12)
plt.ylabel('Passageiros', fontsize=12)

# GRÁFICO 2: Boxplot Vertical com Quartis e Outliers
plt.subplot(1, 2, 2)
boxplot = sns.boxplot(y=df['Age'], color='lightgreen', width=0.5)

# Calcular quartis e outliers
quartis = df['Age'].quantile([0.25, 0.5, 0.75])
Q1 = quartis[0.25]
Q2 = quartis[0.5]  # Mediana
Q3 = quartis[0.75]
IQR = Q3 - Q1  # Intervalo Interquartil

# Definir limites para outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificar outliers
outliers = df[(df['Age'] < limite_inferior) | (df['Age'] > limite_superior)]['Age']

# Adicionar rótulos para quartis
plt.text(0.1, Q1, f'Q1: {Q1:.1f}', color='blue', fontsize=10, ha='center')
plt.text(0.1, Q2, f'Q2 (Mediana): {Q2:.1f}', color='red', fontsize=10, ha='center')
plt.text(0.1, Q3, f'Q3: {Q3:.1f}', color='blue', fontsize=10, ha='center')

# Adicionar rótulos para outliers
if not outliers.empty:
    for i, outlier in enumerate(outliers):
        plt.text(0.1, outlier, f'Outlier: {outlier:.1f}', color='orange', fontsize=9, ha='center')

# Título e rótulos
plt.title('Boxplot', fontsize=14)
plt.ylabel('Idade', fontsize=12)

# Ajustar layout para evitar sobreposições
plt.tight_layout()

# Exibir os gráficos
plt.show()
```

A maioria dos passageiros está concentrada nas faixas etárias centrais, com um pico ao redor das idades médias.
  
Faixas Etárias com Maior Número de Passageiros:

As faixas etárias entre aproximadamente 20 e 60 anos têm o maior número de passageiros, com picos notáveis ao redor das idades de 30 e 50 anos.

### 4.2 Respondendo as Perguntas de Negócio:

**A)Quais perfis demográficos de passageiros estão mais associados a voos com atraso ou cancelamento?**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ================== FUNÇÕES AUXILIARES ==================

def preparar_variavel_numerica(df, feature, bins=5):
    """Agrupa variáveis numéricas em faixas usando quantis."""
    if pd.api.types.is_numeric_dtype(df[feature]):
        new_col = f"{feature}_bin"
        df[new_col] = pd.qcut(df[feature], q=bins, duplicates='drop').astype(str)
        return new_col
    return feature

def calcular_woe_iv(df, feature, target, woe_tables, min_freq=10):
    """Calcula Weight of Evidence (WoE) e Information Value (IV), filtrando categorias raras."""
    try:
        # Verificar valores nulos
        if df[feature].isnull().any():
            print(f"Aviso: {feature} contém valores nulos. Preenchendo com 'Missing'.")
            df[feature] = df[feature].fillna('Missing')
        
        # Criar tabela cruzada
        cross_tab = pd.crosstab(df[feature], df[target], margins=False)
        
        # Filtrar categorias com frequência mínima
        cross_tab['Total'] = cross_tab.sum(axis=1)
        cross_tab = cross_tab[cross_tab['Total'] >= min_freq].drop(columns='Total')
        
        # Verificar se a tabela cruzada está vazia
        if cross_tab.empty:
            print(f"Erro: Tabela cruzada vazia para {feature} após filtragem de frequência")
            return 0
        
        # Adicionar pequena constante para evitar divisão por zero
        cross_tab = cross_tab.replace(0, 0.5)
        
        # Calcular distribuições
        total_0 = cross_tab[0].sum()  # On Time
        total_1 = cross_tab[1].sum()  # Delayed/Cancelled
        
        cross_tab['%_OnTime'] = cross_tab[0] / total_0
        cross_tab['%_BadExperience'] = cross_tab[1] / total_1
        
        # Calcular WoE e IV
        cross_tab['WoE'] = np.log(cross_tab['%_OnTime'] / cross_tab['%_BadExperience'])
        cross_tab['IV'] = (cross_tab['%_OnTime'] - cross_tab['%_BadExperience']) * cross_tab['WoE']
        
        # Formatar para exibição
        cross_tab['%_OnTime'] = (cross_tab['%_OnTime'] * 100).round(1).astype(str) + '%'
        cross_tab['%_BadExperience'] = (cross_tab['%_BadExperience'] * 100).round(1).astype(str) + '%'
        
        # Armazenar resultados
        woe_tables[feature] = cross_tab.copy()
        return cross_tab['IV'].sum()
    
    except Exception as e:
        print(f"Erro ao processar {feature}: {str(e)}")
        return 0

def classificar_poder_preditivo(iv_value):
    """Classifica o poder preditivo com base no IV."""
    if iv_value < 0.02: return 'Sem poder'
    elif iv_value < 0.1: return 'Fraco'
    elif iv_value < 0.3: return 'Moderado'
    elif iv_value < 0.5: return 'Forte'
    else: return 'Muito forte'

def plot_woe_negative(woe_table, feature_name, top_n=20):
    """Gera gráfico de barras para categorias com WoE negativo."""
    woe_negative = woe_table[woe_table['WoE'] < 0][['WoE']].sort_values('WoE', ascending=True).head(top_n)
    if woe_negative.empty:
        print(f"Nenhuma categoria com WoE negativo para {feature_name}")
        return
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='WoE', y=woe_negative.index, data=woe_negative, palette='Reds_r')
    plt.title(f'WoE Negativo - {feature_name} (Maior Risco de Atraso/Cancelamento)', pad=20)
    plt.xlabel('Weight of Evidence (WoE)')
    plt.ylabel('Categorias')
    plt.tight_layout()
    plt.savefig(f'woe_negative_{feature_name}.png')
    plt.show()

def analyze_demographic_profiles(df, airports, demographic_vars, limit=50):
    """Analisa perfis demográficos para voos com aeroportos de alto risco, limitando a saída."""
    profiles = {}
    for airport in airports:
        print(f"\nAnalisando perfis demográficos para Arrival Airport: {airport}")
        print("-"*60)
        airport_df = df[df['Arrival Airport'] == airport]
        profile = {}
        for var in demographic_vars:
            if var in airport_df.columns:
                distribution = airport_df[var].value_counts(normalize=True) * 100
                distribution = distribution.round(1).astype(str) + '%'
                profile[var] = distribution.to_dict()
                print(f"\nDistribuição de {var} (até {limit} categorias):")
                # Limitar a exibição a 50 categorias
                limited_distribution = distribution.head(limit)
                print(limited_distribution.to_string())
                if len(distribution) > limit:
                    print(f"... e mais {len(distribution) - limit} categorias omitidas.")
        profiles[airport] = profile
    return profiles

# ================== EXECUÇÃO PRINCIPAL ==================

# 1. Carregar e preparar os dados
print("⏳ Carregando e preparando dados...")
try:
    df = pd.read_csv("Airline Dataset Updated - v2.csv")
    # Criar variável alvo: 1=Delayed/Cancelled, 0=On Time
    df['Bad_Experience'] = (df['Flight Status'].isin(['Delayed', 'Cancelled'])).astype(int)
except Exception as e:
    print(f"Erro ao carregar dados: {e}")
    raise

# Verificar desbalanceamento em Flight Status
print("\nDistribuição de Flight Status:")
print(df['Flight Status'].value_counts(normalize=True) * 100)

# Verificar valores nulos nas colunas relevantes
print("\nValores nulos nas colunas:")
print(df[['Gender', 'Age', 'Nationality', 'Airport Continent', 'Continents', 'Arrival Airport', 'Departure Date', 'Flight Status']].isnull().sum())

# Pré-processamento
# Corrigir formato de Departure Date (DD-MM-YYYY)
try:
    df['Departure Date'] = pd.to_datetime(df['Departure Date'], format='%d-%m-%Y', errors='coerce')
except Exception as e:
    print(f"Erro na conversão de datas: {e}")
    df['Departure Date'] = pd.to_datetime(df['Departure Date'], dayfirst=True, errors='coerce')

# Extrair mês para sazonalidade
df['Departure_Month'] = df['Departure Date'].dt.month.astype(str)

# Discretizar Age
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+']).astype(str)

# Agrupar Nationalities com baixa frequência
nationality_counts = df['Nationality'].value_counts()
df['Nationality_Grouped'] = df['Nationality'].apply(lambda x: x if nationality_counts[x] > 1000 else 'Other')

# 2. Selecionar e preparar features
features = [
    'Gender',
    'Age_Group',
    'Nationality_Grouped',
    'Airport Continent',
    'Continents',
    'Departure_Month',
    'Arrival Airport'
]

demographic_vars = [
    'Gender',
    'Age_Group',
    'Nationality_Grouped',
    'Airport Continent',
    'Continents',
    'Departure_Month'
]

# Dicionário para mapear nomes originais para nomes processados
processed_features = {}

# Pré-processamento das features
for col in features:
    try:
        processed_name = preparar_variavel_numerica(df, col)
        processed_features[col] = processed_name
    except Exception as e:
        print(f"Erro ao processar {col}: {e}")
        continue

# 3. Calcular WoE e IV
print("📊 Calculando WoE e IV para cada variável...")
woe_results = {}
iv_values = {}

for original_name in features:
    feature_name = processed_features.get(original_name, original_name)
    try:
        if feature_name in df.columns:
            iv = calcular_woe_iv(df, feature_name, 'Bad_Experience', woe_results, min_freq=10)
            iv_values[original_name] = iv
        else:
            print(f"Variável {feature_name} não encontrada no DataFrame")
            iv_values[original_name] = 0
    except Exception as e:
        print(f"Erro ao calcular IV para {original_name}: {e}")
        iv_values[original_name] = 0

# 4. Criar tabela de resultados
iv_df = pd.DataFrame.from_dict(iv_values, orient='index', columns=['IV'])
iv_df['Poder Preditivo'] = iv_df['IV'].apply(classificar_poder_preditivo)
iv_df = iv_df.sort_values('IV', ascending=False)

# Exibir tabela de IV para todas as variáveis
print("\n📋 Tabela de Information Value (IV) para Todas as Variáveis")
print("="*60)
print(iv_df.to_string())
print("\nResultados do IV salvos em 'iv_analysis_results.csv'")
iv_df.to_csv('iv_analysis_results.csv')

# ================== VISUALIZAÇÃO ==================

# Gráfico de IV
plt.figure(figsize=(10,6))
ax = sns.barplot(x='IV', y=iv_df.index, data=iv_df, 
                 hue=iv_df.index, palette='viridis', dodge=False, legend=False)

# Adicionar valores nas barras
for p in ax.patches:
    ax.annotate(f"{p.get_width():.3f}", 
                (p.get_width() + 0.01, p.get_y() + p.get_height()/2),
                ha='left', va='center')

plt.title('Information Value por Variável (Risco de Atraso/Cancelamento)', pad=20)
plt.xlabel('Information Value (IV)')
plt.ylabel('Variáveis')
plt.tight_layout()
plt.savefig('iv_plot.png')
plt.show()

# ================== RESULTADOS DETALHADOS ==================

# 5. Exibir tabelas para todas as variáveis
print("\n🔍 RELATÓRIO DETALHADO PARA TODAS AS VARIÁVEIS")
print("="*85)

high_risk_profiles = []
limit = 50  # Definir limite para exibição

for original_name in iv_df.index:
    feature_name = processed_features.get(original_name, original_name)
    
    if feature_name not in woe_results:
        print(f"\n⚠️ Dados não disponíveis para: {original_name}")
        continue
        
    print(f"\n► Variável: {original_name} (IV = {iv_df.loc[original_name,'IV']:.3f} - {iv_df.loc[original_name,'Poder Preditivo']})")
    print("-"*60)
    
    # Preparar tabela WoE
    try:
        woe_table = woe_results[feature_name].copy()
        cols_to_show = ['%_OnTime', '%_BadExperience', 'WoE', 'IV']
        
        # Verificar se as colunas existem
        available_cols = [col for col in cols_to_show if col in woe_table.columns]
        if not available_cols:
            print("Colunas esperadas não encontradas na tabela WoE")
            continue
            
        woe_table = woe_table[available_cols]
        
        # Arredondar valores numéricos
        if 'WoE' in woe_table.columns:
            woe_table['WoE'] = woe_table['WoE'].round(3)
        if 'IV' in woe_table.columns:
            woe_table['IV'] = woe_table['IV'].round(5)
        
        # Exibir até 15 categorias para variáveis demográficas, 50 para Arrival Airport
        max_categories = limit if original_name == 'Arrival Airport' else 15
        print(f"\nTabela WoE (Primeiras {max_categories} Categorias):")
        limited_woe = woe_table.sort_values('WoE', ascending=False).head(max_categories)
        print(limited_woe.to_string())
        if len(woe_table) > max_categories:
            print(f"... e mais {len(woe_table) - max_categories} categorias omitidas.")
        
        # Exibir categorias com WoE negativo (associadas a atraso/cancelamento)
        print("\nCategorias com Maior Risco (WoE Negativo):")
        woe_negative = woe_table[woe_table['WoE'] < 0]
        if not woe_negative.empty:
            limited_negative = woe_negative.sort_values('WoE', ascending=True).head(limit)
            print(limited_negative.to_string())
            if len(woe_negative) > limit:
                print(f"... e mais {len(woe_negative) - limit} categorias omitidas.")
            # Armazenar perfis de alto risco
            for category in woe_negative.index:
                high_risk_profiles.append(f"{original_name}: {category} (WoE = {woe_negative.loc[category, 'WoE']:.3f})")
        else:
            print("Nenhuma categoria com WoE negativo encontrada.")
        
        # Exibir tabela WoE completa, limitada a 50
        print(f"\nTabela WoE Completa (até {limit} categorias):")
        limited_full_woe = woe_table.sort_values('WoE', ascending=False).head(limit)
        print(limited_full_woe.to_string())
        if len(woe_table) > limit:
            print(f"... e mais {len(woe_table) - limit} categorias omitidas.")
        
        # Interpretação
        if 'WoE' in woe_table.columns:
            top_ontime = woe_table['WoE'].idxmax()
            top_badexperience = woe_table['WoE'].idxmin()
            print(f"\nInterpretação:")
            print(f"- Maior associação com PONTUALIDADE: {top_ontime} (WoE = {woe_table.loc[top_ontime,'WoE']:.2f})")
            print(f"- Maior associação com ATRASO/CANCELAMENTO: {top_badexperience} (WoE = {woe_table.loc[top_badexperience,'WoE']:.2f})")
        else:
            print("\nNão foi possível gerar interpretação (WoE não calculado)")
        
        # Gerar gráfico de WoE negativo
        print(f"\nGerando gráfico de WoE Negativo para {original_name}...")
        plot_woe_negative(woe_table, original_name, top_n=20 if original_name == 'Arrival Airport' else 15)
            
    except Exception as e:
        print(f"Erro ao processar tabela WoE: {e}")
    
    print("="*85)

# 6. Segmentação por Arrival Airport
print("\n🔍 SEGMENTAÇÃO DEMOGRÁFICA PARA AEROPORTOS DE ALTO RISCO")
print("="*85)

# Identificar aeroportos de alto risco (WoE negativo, top 10)
if 'Arrival Airport' in woe_results:
    woe_arrival = woe_results['Arrival Airport']
    high_risk_airports = woe_arrival[woe_arrival['WoE'] < 0].index.tolist()
    high_risk_airports = high_risk_airports[:10]  # Limitar a 10 para análise
    print(f"\nAeroportos de alto risco selecionados (Top 10 com WoE Negativo): {high_risk_airports}")
    
    # Analisar perfis demográficos
    profiles = analyze_demographic_profiles(df, high_risk_airports, demographic_vars, limit=50)
    
    # Salvar perfis em CSV
    profile_data = []
    for airport, profile in profiles.items():
        for var, dist in profile.items():
            for category, percentage in dist.items():
                profile_data.append({
                    'Arrival Airport': airport,
                    'Variable': var,
                    'Category': category,
                    'Percentage': percentage
                })
    profile_df = pd.DataFrame(profile_data)
    profile_df.to_csv('demographic_profiles_high_risk_airports.csv', index=False)
    print("\nPerfis demográficos salvos em 'demographic_profiles_high_risk_airports.csv'")
else:
    print("\n⚠️ Dados de Arrival Airport não disponíveis para segmentação")

# 7. Resumo dos perfis de maior risco
print("\n📌 RESUMO DOS PERFIS DE MAIOR RISCO")
print("="*60)
if high_risk_profiles:
    print("Perfis mais expostos a atrasos/cancelamentos:")
    for profile in high_risk_profiles[:50]:  # Limitar a 50 perfis
        print(f"- {profile}")
    if len(high_risk_profiles) > 50:
        print(f"... e mais {len(high_risk_profiles) - 50} perfis omitidos.")
else:
    print("Nenhum perfil com WoE negativo identificado.")
print("\nVerifique 'iv_analysis_results.csv', 'demographic_profiles_high_risk_airports.csv', e gráficos 'woe_negative_*.png' para mais detalhes.")
```

O cálculo do Information Value (IV), uma técnica estatística utilizada para avaliar o poder preditivo de variáveis independentes em relação a uma variável de resultado.

Quanto maior o IV de uma variável, maior é a sua capacidade de discriminar entre as categorias da variável resultado.

**Explicação da Tecnica**

Random Forest (Floresta Aleatória) é um algoritmo de aprendizado de máquina que pertence à família dos métodos ensemble (que combinam múltiplos modelos).

Ele funciona criando muitas árvores de decisão durante o treinamento e produzindo como resultado a moda (classificação) ou média (regressão) das previsões das árvores individuais.


Como interpretar os resultados do Random Forest?

*Interpretação:*

Valores entre 0 e 1: Quanto maior, mais importante a feature é para prever o atraso

Soma total 1: A importância é relativa entre as features

Airport Country Code (44%) e Country Name (43%) são os fatores mais determinantes

Continentes têm influência menor (~6% cada)

Por que usamos Random Forest neste contexto?


Identificação de padrões complexos: Relações não-lineares entre países/continentes e atrasos

Ranking de importância: Saber quais fatores geográficos impactam mais os atrasos

Robustez a outliers: Menos sensível a dados anômalos que modelos paramétricos

Tratamento de variáveis categóricas: Lida bem com códigos de países/nomes sem necessidade de pré-processamento complexo

Interpretabilidade: Permite extrair insights mesmo sendo um modelo sofisticado

  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import warnings

# Configurações para suprimir warnings
pd.options.mode.chained_assignment = None
warnings.filterwarnings('ignore')

# Carregar os dados
df = pd.read_csv("Airline Dataset Updated - v2.csv")

# 1. Análise de Atrasos por Continente
plt.figure(figsize=(10, 6))
atrasados = df[df['Flight Status'] == 'Delayed'].copy()
sns.countplot(data=atrasados, x='Airport Continent', 
             order=atrasados['Airport Continent'].value_counts().index)
plt.title('Atrasos por Continente do aeroporto de origem')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Top 10 Países com mais Atrasos
plt.figure(figsize=(12, 6))
top_paises = atrasados['Country Name'].value_counts().head(10)
sns.barplot(x=top_paises.values, y=top_paises.index)
plt.title('Top 10 países com mais atrasos')
plt.xlabel('Número de atrasos')
plt.tight_layout()
plt.show()

# 3. Top 10 Rotas com mais Atrasos
atrasados.loc[:, 'Rota'] = atrasados['Airport Name'] + ' - ' + atrasados['Arrival Airport']
top_rotas = atrasados['Rota'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_rotas.values, y=top_rotas.index)
plt.title('Top 10 rotas com mais atrasos')
plt.xlabel('Número de atrasos')
plt.tight_layout()
plt.show()

# 4. Análise com Random Forest para identificar fatores importantes
# Preparar dados
df_model = df.copy()

# Filtrar apenas status de voo que fazem sentido para classificação
status_validos = ['Delayed', 'On Time', 'Cancelled']  # ajuste conforme seus dados
df_model = df_model[df_model['Flight Status'].isin(status_validos)].copy()

# Codificar variável target
le_status = LabelEncoder()
df_model.loc[:, 'Flight Status_encoded'] = le_status.fit_transform(df_model['Flight Status'])

# Codificar variáveis categóricas
categorical_cols = ['Airport Country Code', 'Country Name', 'Airport Continent', 'Continents']
le_features = LabelEncoder()

for col in categorical_cols:
    df_model.loc[:, col+'_encoded'] = le_features.fit_transform(df_model[col].astype(str))

# Treinar modelo
X = df_model[[col+'_encoded' for col in categorical_cols]]
y = df_model['Flight Status_encoded']

if len(y.unique()) > 1:  # Verificar se temos pelo menos duas classes
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # Mostrar importância das features
    feature_importance = pd.DataFrame({
        'Feature': categorical_cols,
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)
    
    print("\nFatores mais importantes para prever atrasos:")
    print(feature_importance)
    
    # Gráfico de importância
    plt.figure(figsize=(10, 5))
    sns.barplot(data=feature_importance, x='Importance', y='Feature')
    plt.title('Importância dos fatores geográficos nos atrasos')
    plt.tight_layout()
    plt.show()
else:
    print("\nAviso: Não há variação suficiente na variável target para treinar o modelo.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import warnings

# Configurações iniciais
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", palette="muted")

# 1. Carregar e verificar os dados
try:
    df = pd.read_csv("Airline Dataset Updated - v2.csv")
    
    # Verificar colunas existentes
    required_columns = ['Pilot Name', 'Flight Status', 'Airport Country Code', 'Airport Continent', 'Gender', 'Age']
    missing_cols = [col for col in required_columns if col not in df.columns]
    
    if missing_cols:
        raise KeyError(f"Colunas faltantes no dataset: {missing_cols}")
        
except Exception as e:
    print(f"Erro ao carregar dados: {str(e)}")
    exit()

# 2. Pré-processamento
# Definir voos com problemas (ajuste conforme seus status reais)
status_problema = ['Delayed', 'Cancelled', 'Diverted']
df['Problema'] = df['Flight Status'].isin(status_problema).astype(int)

# Verificar distribuição
print("\nDistribuição de status de voo:")
print(df['Flight Status'].value_counts(normalize=True))

# 3. Análise de pilotos
# Calcular estatísticas por piloto
pilotos_stats = df.groupby('Pilot Name').agg(
    Total_Voos=('Flight Status', 'count'),
    Voos_Problema=('Problema', 'sum')
).reset_index()

# Calcular taxa e filtrar pilotos com suficiente experiência
pilotos_stats['Taxa_Problema'] = pilotos_stats['Voos_Problema'] / pilotos_stats['Total_Voos']
pilotos_signif = pilotos_stats[pilotos_stats['Total_Voos'] >= 10].sort_values('Taxa_Problema', ascending=False)

# Visualização
plt.figure(figsize=(12, 8))
if not pilotos_signif.empty:
    top_pilotos = pilotos_signif.head(15)
    sns.barplot(data=top_pilotos, y='Pilot Name', x='Taxa_Problema', palette='Reds_r')
    plt.title('Top 15 Pilotos com Maior Taxa de Problemas (min. 10 voos)')
    plt.xlabel('Taxa de Problemas')
    plt.ylabel('')
    plt.xlim(0, 1)
    plt.tight_layout()
    plt.show()
else:
    print("\nNão há pilotos com número suficiente de voos para análise.")

# 4. Análise de grupos operacionais
if all(col in df.columns for col in ['Pilot Name', 'Airport Country Code']):
    df['Grupo_Operacional'] = df['Pilot Name'] + ' | ' + df['Airport Country Code']
    grupos_stats = df.groupby('Grupo_Operacional').agg(
        Total_Voos=('Flight Status', 'count'),
        Voos_Problema=('Problema', 'sum')
    ).reset_index()
    
    grupos_stats['Taxa_Problema'] = grupos_stats['Voos_Problema'] / grupos_stats['Total_Voos']
    grupos_signif = grupos_stats[grupos_stats['Total_Voos'] >= 5].sort_values('Taxa_Problema', ascending=False)
    
    plt.figure(figsize=(12, 8))
    if not grupos_signif.empty:
        top_grupos = grupos_signif.head(15)
        sns.barplot(data=top_grupos, y='Grupo_Operacional', x='Taxa_Problema', palette='OrRd_r')
        plt.title('Top 15 Grupos Operacionais com Problemas (min. 5 voos)')
        plt.xlabel('Taxa de Problemas')
        plt.ylabel('Piloto | País Aeroporto')
        plt.xlim(0, 1)
        plt.tight_layout()
        plt.show()
    else:
        print("\nNão há grupos com número suficiente de voos para análise.")
else:
    print("\nColunas necessárias para análise de grupos não encontradas.")

# 5. Modelagem com Random Forest
print("\nPreparando modelo de Random Forest...")
try:
    # Codificação de variáveis
    le = LabelEncoder()
    df_model = df.copy()
    
    # Codificar variáveis categóricas
    cat_cols = ['Pilot Name', 'Gender', 'Airport Country Code', 'Airport Continent']
    for col in cat_cols:
        if col in df_model.columns:
            df_model[col+'_enc'] = le.fit_transform(df_model[col].astype(str))
    
    # Selecionar features
    features = [col+'_enc' for col in cat_cols if col+'_enc' in df_model.columns]
    if 'Age' in df_model.columns:
        features.append('Age')
    
    if not features:
        raise ValueError("Nenhuma feature disponível para o modelo")
    
    X = df_model[features]
    y = df_model['Problema']
    
    # Treinar modelo
    rf = RandomForestClassifier(n_estimators=150, random_state=42, class_weight='balanced', max_depth=5)
    rf.fit(X, y)
    
    # Importância das features
    feature_importance = pd.DataFrame({
        'Feature': [col.replace('_enc', '') for col in features],
        'Importance': rf.feature_importances_
    }).sort_values('Importance', ascending=False)
    
    print("\nImportância dos Fatores na Previsão de Problemas:")
    print(feature_importance)
    
    # Plotar importância
    plt.figure(figsize=(10, 5))
    sns.barplot(data=feature_importance, y='Feature', x='Importance', palette='Blues_r')
    plt.title('Importância dos Fatores no Modelo Random Forest')
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"\nErro na modelagem: {str(e)}")
```

**Explicação**


*O que é o teste qui-quadrado?*


O teste qui-quadrado (χ²) é um teste estatístico usado para determinar se existe uma associação significativa entre duas variáveis categóricas. No contexto da sua análise, essas variáveis são:

Variáveis temporais (dia da semana, período do dia, mês)


Status do voo (atrasado ou não atrasado)

*O que o teste faz neste código?*

No código, o teste qui-quadrado está sendo aplicado em três análises diferentes:


python# Teste qui-quadrado para dia da semana


contingency_day = pd.crosstab(df['Departure_DayOfWeek'], df['Atraso'])

chi2_day, p_day, dof_day, expected_day = stats.chi2_contingency(contingency_day)

# Teste qui-quadrado para período do dia


contingency_time = pd.crosstab(df['Departure_TimeCategory'], df['Atraso'])

chi2_time, p_time, dof_time, expected_time = stats.chi2_contingency(contingency_time)

# Teste qui-quadrado para mês

contingency_month = pd.crosstab(df['Departure_Month'], df['Atraso'])

chi2_month, p_month, dof_month, expected_month = stats.chi2_contingency(contingency_month)

*Cada teste:*

Cria uma tabela de contingência (crosstab) entre a variável temporal e o atraso

Calcula a estatística qui-quadrado e o valor-p


Avalia se existe diferença estatisticamente significativa

*Como interpretar os resultados?*

O valor mais importante para interpretar é o p-valor (p_day, p_time, p_month):

Se p-valor < 0.05: Existe uma associação estatisticamente significativa. Isso significa que as variações nos atrasos por aquela dimensão temporal (dia, horário ou mês) NÃO são aleatórias.

Se p-valor ≥ 0.05: Não há evidência suficiente para afirmar que existe uma associação. As variações observadas podem ser resultado do acaso.

*Por exemplo, quando o código imprime:*

Teste qui-quadrado para dia da semana: chi2=15.27, p-valor=0.0183

Diferença estatisticamente significativa? Sim

Isso significa que os atrasos variam significativamente dependendo do dia da semana, e essa variação não é apenas aleatória.

*Por que foi necessário usar o teste qui-quadrado?*

Confirmação estatística: Não basta observar que as taxas de atraso são diferentes (por exemplo, 30% na segunda-feira vs. 25% na terça-feira). É preciso confirmar que essas diferenças são estatisticamente significativas.


Evitar conclusões errôneas: Sem um teste estatístico, poderíamos interpretar variações aleatórias como padrões reais. Por exemplo, se tivermos poucos voos na madrugada, a taxa de atraso pode parecer alta ou baixa por acaso.


Responder objetivamente à pergunta: A questão "O padrão de atrasos varia por horário ou data específica?" requer uma resposta sólida baseada em evidências estatísticas, não apenas em observações visuais dos gráficos.


Base para decisões operacionais: Se a companhia aérea quiser implementar mudanças operacionais baseadas nessa análise (como aumentar a equipe em determinados horários), precisa ter certeza de que está respondendo a um padrão real.

Em resumo, o teste qui-quadrado fornece o rigor estatístico necessário para transformar sua análise de dados em conclusões acionáveis sobre os padrões de atraso. Sem ele, teria apenas descrições dos dados, mas não poderia afirmar com confiança se existe um padrão temporal significativo nos atrasos.





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import matplotlib.dates as mdates

# Carregar o dataset
df = pd.read_csv("Airline Dataset Updated - v2.csv")

# Converter Departure Date para datetime com o formato correto (dia-mês-ano)
df['Departure_DateTime'] = pd.to_datetime(df['Departure Date'], format='%d-%m-%Y', errors='coerce')

# Verificar se há valores nulos após a conversão
print("Valores nulos na data de partida:", df['Departure_DateTime'].isnull().sum())

# Remover linhas com datas inválidas, se houver
df = df.dropna(subset=['Departure_DateTime'])

# Extrair features temporais da data
df['Departure_Hour'] = df['Departure_DateTime'].dt.hour
df['Departure_Day'] = df['Departure_DateTime'].dt.day
df['Departure_Month'] = df['Departure_DateTime'].dt.month
df['Departure_Year'] = df['Departure_DateTime'].dt.year
df['Departure_DayOfWeek'] = df['Departure_DateTime'].dt.dayofweek  # 0=Segunda, 6=Domingo
df['Departure_Weekend'] = df['Departure_DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)

# Mapear nomes dos dias da semana
day_names = {0: 'Segunda', 1: 'Terça', 2: 'Quarta', 3: 'Quinta', 
             4: 'Sexta', 5: 'Sábado', 6: 'Domingo'}
df['Departure_DayName'] = df['Departure_DayOfWeek'].map(day_names)

# Verificar se há informação de hora na coluna original
print("Valores únicos nas horas de partida:", df['Departure_Hour'].unique())

# Se não houver informação de hora (todas NaN ou valor único), consideramos hora=0
if df['Departure_Hour'].isnull().all() or len(df['Departure_Hour'].unique()) == 1:
    print("AVISO: Não há informação de hora variada no dataset. Utilizando valores simulados para demonstração.")
    # Simular horas para demonstração (remover em produção)
    df['Departure_Hour'] = np.random.randint(0, 24, size=len(df))

# Criar categorias de horário
def categorize_hour(hour):
    if 5 <= hour < 12:
        return 'Manhã'
    elif 12 <= hour < 18:
        return 'Tarde'
    elif 18 <= hour < 24:
        return 'Noite'
    else:
        return 'Madrugada'

df['Departure_TimeCategory'] = df['Departure_Hour'].apply(categorize_hour)

# Verificar valores únicos em Flight Status
print("Valores únicos em Flight Status:", df['Flight Status'].unique())

# Criar coluna de atraso (ajuste conforme necessário para seu dataset)
df['Atraso'] = df['Flight Status'].apply(lambda x: 1 if 'delayed' in str(x).lower() or 'atrasado' in str(x).lower() else 0)

print("\n=== ANÁLISE: O PADRÃO DE ATRASOS VARIA POR HORÁRIO OU DATA ESPECÍFICA? ===")

# 1. Análise por hora do dia
atrasos_hora = df.groupby('Departure_Hour')['Atraso'].mean()
print("\n1. Variação de atrasos por hora do dia:")
print(atrasos_hora)

# 2. Análise por categoria de horário
atrasos_categoria = df.groupby('Departure_TimeCategory')['Atraso'].agg(['mean', 'count'])
print("\n2. Variação de atrasos por período do dia:")
print(atrasos_categoria)

# 3. Análise por dia da semana
atrasos_dia_semana = df.groupby('Departure_DayName')['Atraso'].mean()
dia_semana_order = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']
atrasos_dia_semana = atrasos_dia_semana.reindex(dia_semana_order)
print("\n3. Variação de atrasos por dia da semana:")
print(atrasos_dia_semana)

# 4. Análise por mês
atrasos_mes = df.groupby('Departure_Month')['Atraso'].mean()
print("\n4. Variação de atrasos por mês:")
print(atrasos_mes)

# 5. Análise por dia do mês
atrasos_dia = df.groupby('Departure_Day')['Atraso'].mean()
print("\n5. Variação de atrasos por dia do mês:")
print(atrasos_dia)

# 6. Teste estatístico para verificar significância das diferenças
from scipy import stats

# Teste qui-quadrado para dia da semana
print("\n6. Testes estatísticos:")
contingency_day = pd.crosstab(df['Departure_DayOfWeek'], df['Atraso'])
chi2_day, p_day, dof_day, expected_day = stats.chi2_contingency(contingency_day)
print(f"Teste qui-quadrado para dia da semana: chi2={chi2_day:.2f}, p-valor={p_day:.4f}")
print(f"Diferença estatisticamente significativa? {'Sim' if p_day < 0.05 else 'Não'}")

# Teste qui-quadrado para período do dia
contingency_time = pd.crosstab(df['Departure_TimeCategory'], df['Atraso'])
chi2_time, p_time, dof_time, expected_time = stats.chi2_contingency(contingency_time)
print(f"Teste qui-quadrado para período do dia: chi2={chi2_time:.2f}, p-valor={p_time:.4f}")
print(f"Diferença estatisticamente significativa? {'Sim' if p_time < 0.05 else 'Não'}")

# Teste qui-quadrado para mês
contingency_month = pd.crosstab(df['Departure_Month'], df['Atraso'])
chi2_month, p_month, dof_month, expected_month = stats.chi2_contingency(contingency_month)
print(f"Teste qui-quadrado para mês: chi2={chi2_month:.2f}, p-valor={p_month:.4f}")
print(f"Diferença estatisticamente significativa? {'Sim' if p_month < 0.05 else 'Não'}")

# Visualizações
plt.figure(figsize=(20, 15))

# Gráfico 1: Atrasos por hora do dia (gráfico de linha)
plt.subplot(3, 2, 1)
atrasos_hora.plot(kind='line', marker='o')
plt.title('Taxa de Atrasos por Hora do Dia')
plt.xlabel('Hora do Dia')
plt.ylabel('Taxa de Atrasos')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.7)

# Gráfico 2: Atrasos por categoria de horário
plt.subplot(3, 2, 2)
sns.barplot(x=atrasos_categoria.index, y=atrasos_categoria['mean'], 
            order=['Madrugada', 'Manhã', 'Tarde', 'Noite'])
plt.title('Taxa de Atrasos por Período do Dia')
plt.ylabel('Taxa de Atrasos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 3: Atrasos por dia da semana
plt.subplot(3, 2, 3)
sns.barplot(x=atrasos_dia_semana.index, y=atrasos_dia_semana.values)
plt.title('Taxa de Atrasos por Dia da Semana')
plt.xticks(rotation=45)
plt.ylabel('Taxa de Atrasos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 4: Atrasos por mês (heatmap)
plt.subplot(3, 2, 4)
month_names = {1: 'Jan', 2: 'Fev', 3: 'Mar', 4: 'Abr', 5: 'Mai', 6: 'Jun',
               7: 'Jul', 8: 'Ago', 9: 'Set', 10: 'Out', 11: 'Nov', 12: 'Dez'}
atrasos_mes.index = atrasos_mes.index.map(lambda x: month_names.get(x, x))
sns.barplot(x=atrasos_mes.index, y=atrasos_mes.values)
plt.title('Taxa de Atrasos por Mês')
plt.ylabel('Taxa de Atrasos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 5: Atrasos por dia do mês
plt.subplot(3, 2, 5)
plt.plot(atrasos_dia.index, atrasos_dia.values, marker='o', linestyle='-', color='purple')
plt.title('Taxa de Atrasos por Dia do Mês')
plt.xlabel('Dia do Mês')
plt.ylabel('Taxa de Atrasos')
plt.xticks(range(1, 32, 2))
plt.grid(True, linestyle='--', alpha=0.7)

# Gráfico 6: Box plot comparando dias da semana
plt.subplot(3, 2, 6)
sns.boxplot(x='Departure_DayName', y='Atraso', data=df, order=dia_semana_order)
plt.title('Distribuição de Atrasos por Dia da Semana')
plt.xticks(rotation=45)
plt.ylabel('Taxa de Atrasos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('analise_atrasos_padrao_temporal.png', dpi=300, bbox_inches='tight')
plt.show()

# Mapa de calor de horário por dia da semana
plt.figure(figsize=(12, 8))
# Criar uma tabela pivot com dias da semana x horários
pivot_day_hour = df.pivot_table(
    values='Atraso', 
    index='Departure_DayName', 
    columns='Departure_Hour',
    aggfunc='mean'
)
# Reordenar os dias da semana
pivot_day_hour = pivot_day_hour.reindex(dia_semana_order)
# Plotar o mapa de calor
sns.heatmap(pivot_day_hour, cmap="YlOrRd", annot=True, fmt=".2f", cbar_kws={'label': 'Taxa de Atrasos'})
plt.title('Mapa de Calor: Taxa de Atrasos por Dia da Semana e Hora')
plt.xlabel('Hora do Dia')
plt.ylabel('Dia da Semana')
plt.savefig('mapa_calor_atrasos.png', dpi=300, bbox_inches='tight')
plt.show()

# Conclusão
print("\n=== CONCLUSÃO: O PADRÃO DE ATRASOS VARIA POR HORÁRIO OU DATA ESPECÍFICA? ===")
print("1. Por horário:", "Sim" if p_time < 0.05 else "Não")
print("2. Por dia da semana:", "Sim" if p_day < 0.05 else "Não")
print("3. Por mês:", "Sim" if p_month < 0.05 else "Não")

print("\nResumo:")
if p_time < 0.05 or p_day < 0.05 or p_month < 0.05:
    print("Os padrões de atraso VARIAM significativamente por horário e/ou data específica.")
    
    if p_time < 0.05:
        max_cat = atrasos_categoria['mean'].idxmax()
        min_cat = atrasos_categoria['mean'].idxmin()
        print(f"- O período do dia com maior taxa de atrasos é: {max_cat} ({atrasos_categoria.loc[max_cat, 'mean']:.2%})")
        print(f"- O período do dia com menor taxa de atrasos é: {min_cat} ({atrasos_categoria.loc[min_cat, 'mean']:.2%})")
        
    if p_day < 0.05:
        max_day = atrasos_dia_semana.idxmax()
        min_day = atrasos_dia_semana.idxmin()
        print(f"- O dia da semana com maior taxa de atrasos é: {max_day} ({atrasos_dia_semana[max_day]:.2%})")
        print(f"- O dia da semana com menor taxa de atrasos é: {min_day} ({atrasos_dia_semana[min_day]:.2%})")
        
    if p_month < 0.05:
        max_month = atrasos_mes.idxmax()
        min_month = atrasos_mes.idxmin()
        print(f"- O mês com maior taxa de atrasos é: {month_names[max_month]} ({atrasos_mes[max_month]:.2%})")
        print(f"- O mês com menor taxa de atrasos é: {month_names[min_month]} ({atrasos_mes[min_month]:.2%})")
else:
    print("Não há evidência estatística de que os padrões de atraso variem significativamente por horário ou data específica.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.backends.backend_pdf import PdfPages
import webbrowser
import os
import textwrap

def gerar_pdf_5w2h(data):
    # Configuração do documento PDF
    pdf_pages = PdfPages("Plano_Estrategico_AeroVision_5W2H.pdf")

    # Ajustes de layout
    fig, ax = plt.subplots(figsize=(14, 12))  # Aumentei a altura para caber melhor
    ax.axis('off')

    # Configurações de fonte
    font = FontProperties()
    font.set_size(10)
    font_bold = FontProperties()
    font_bold.set_size(11)
    font_bold.set_weight('bold')

    # Preparar os dados com quebras de linha
    wrapped_data = []
    max_lines = 1
    for elemento, descricao in zip(data['Elemento'], data['Descrição']):
        # Aplicar quebra de linha para descrições longas (limite de 70 caracteres por linha)
        wrapped_desc = textwrap.fill(descricao, width=70)
        lines = wrapped_desc.count('\n') + 1
        if lines > max_lines:
            max_lines = lines
        wrapped_data.append([elemento, wrapped_desc])

    # Criação da tabela
    table = ax.table(
        cellText=wrapped_data,
        colLabels=['Elemento', 'Descrição'],
        cellLoc='left',
        loc='center',
        colWidths=[0.15, 0.85]  # Ajuste das larguras das colunas
    )

    # Ajustes de formatação
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1.2, 1.2)  # Ajuste da escala

    # Estilização das células
    for (i, j), cell in table.get_celld().items():
        cell.set_edgecolor('#4682b4')
        cell.set_linewidth(0.7)
        if i == 0:
            cell.set_facecolor('#1e3d59')
            cell.set_text_props(color='white', fontproperties=font_bold)
        else:
            # Ajustar altura das linhas baseado no número de linhas do texto
            cell.set_height(0.15 * max_lines)
            cell.set_text_props(fontproperties=font)
            cell.PAD = 0.1  # Espaçamento interno

    # Finalização do layout
    plt.tight_layout(pad=3.0)
    pdf_pages.savefig(fig, bbox_inches='tight', dpi=300)
    pdf_pages.close()
    return "Plano_Estrategico_AeroVision_5W2H.pdf"

def gerenciar_pdf(pdf_path):
    print("\nOpções de Visualização:")
    resposta = input("[A]brir PDF | [B]aixar PDF | [V]isualizar Tabela: ").upper()

    if resposta == 'A':
        webbrowser.open(pdf_path)
    elif resposta == 'B':
        print(f"\nArquivo disponível em:\n{os.path.abspath(pdf_path)}")
    elif resposta == 'V':
        print("\n" + "=" * 100)
        print(pd.DataFrame(data).to_string(index=False))
        print("=" * 100)
    else:
        print("Opção inválida. Por favor, tente novamente.")

# Dados do plano 5W2H para AeroVision
data = {
    "Elemento": [
        "What (O quê?)",
        "Why (Por quê?)",
        "Where (Onde?)",
        "When (Quando?)",
        "Who (Quem?)",
        "How (Como?)",
        "How Much (Quanto?)"
    ],
    "Descrição": [
        "Realizar uma análise de dados sobre tráfego aéreo nacional para identificar aeroportos estratégicos que possam se beneficiar das soluções analíticas da AeroVision",
        "Compreender o perfil operacional dos aeroportos brasileiros e mapear potenciais clientes com alta demanda por inteligência de dados",
        "Em todo o território nacional, com foco em aeroportos com maior volume e crescimento de operações",
        "Abril de 2025 – Etapa de pesquisa concluída. Próxima etapa: abordagem comercial até junho de 2025",
        "Equipe de Dados (análise), Equipe de Produto (desenvolvimento das soluções), Equipe Comercial (prospecção e venda)",
        "Análise exploratória dos dados públicos de operações aéreas, uso de dashboards interativos e relatórios com recomendações comerciais",
        "Investimento estimado: R$ 35.000 (análise + protótipo de solução + prospecção comercial)"
    ]
}

if __name__ == "__main__":
    df = pd.DataFrame(data)
    arquivo_pdf = gerar_pdf_5w2h(df)
    gerenciar_pdf(arquivo_pdf)
```

## 5.1 Scrum do Projeto

```python
def gerar_pdf_scrum():
    # Nome do arquivo PDF
    pdf_filename = "Plano_Scrum_AeroVision_Analytics.pdf"
    pdf_pages = PdfPages(pdf_filename)

    # Função auxiliar para criar tabelas
    def criar_tabela(ax, data, titulo):
        ax.axis('off')
        ax.set_title(titulo, fontsize=14, fontweight='bold', pad=10)

        df = pd.DataFrame(data)
        table = ax.table(
            cellText=df.values,
            colLabels=list(df.columns),
            cellLoc='left',
            loc='center',
            colWidths=[0.2, 0.8]
        )

        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1.1, 1.5)

        for (i, j), cell in table.get_celld().items():
            cell.set_edgecolor('#3a7ca5')
            if i == 0:
                cell.set_facecolor('#2f6690')
                cell.set_text_props(color='white', weight='bold')
            else:
                cell.set_height(0.2)

    # Criar figura com 4 seções
    fig, axs = plt.subplots(4, 1, figsize=(14, 18))

    # 1. Visão do Produto
    data_visao = {
        "Item": ["Objetivo", "Valor Esperado"],
        "Descrição": [
            "Desenvolver análises estratégicas de tráfego aéreo para aeroportos brasileiros",
            "Fornecer inteligência de dados para otimização operacional e identificação de oportunidades comerciais"
        ]
    }
    criar_tabela(axs[0], data_visao, "Visão do Produto - AeroVision Analytics para Aeroportos")

    # 2. Backlog do Produto (Sprints)
    data_sprints = {
        "Sprint": ["📅 Sprint 1 – Diagnóstico de Mercado (2 semanas)", 
                  "📅 Sprint 2 – Prototipagem de Solução (3 semanas)", 
                  "📅 Sprint 3 – Abordagem Comercial (4 semanas)"],
        "Tarefas": [
            "Explorar e limpar os dados de tráfego aéreo\nResponder às perguntas de negócio\nGerar dashboards e relatórios interativos",
            "Desenvolver visualizações personalizadas por aeroporto\nCriar alertas de variação de tráfego, sazonalidade e crescimento\nCriar apresentação executiva para possíveis clientes",
            "Identificar os 10 aeroportos prioritários\nCustomizar a apresentação para cada cliente\nRealizar reuniões com tomadores de decisão"
        ]
    }
    criar_tabela(axs[1], data_sprints, "Plano de Sprints")

    # 3. Equipe Scrum
    data_equipe = {
        "Função": ["🧑‍💼 Product Owner", 
                  "🧑‍💼 Scrum Master", 
                  "🧑‍💼 Dev Team"],
        "Responsabilidades": [
            "Head de Produto da AeroVision - define prioridades e visão do produto",
            "Analista Líder de Dados - facilita processos ágeis e remove obstáculos",
            "Cientista de Dados, Engenheiro de Dados, Designer de UX, Representante Comercial - desenvolvimento das soluções"
        ]
    }
    criar_tabela(axs[2], data_equipe, "Papéis e Responsabilidades")

    # 4. Métricas de Sucesso
    data_metricas = {
        "Métrica": ["Cobertura de dados", 
                   "Relevância estratégica", 
                   "Qualidade das visualizações", 
                   "Conversão comercial"],
        "Meta": [
            "Dados completos dos principais aeroportos nacionais",
            "Insights que respondam às necessidades reais dos aeroportos",
            "Dashboards intuitivos e acionáveis",
            "Pelo menos 3 propostas comerciais enviadas"
        ]
    }
    criar_tabela(axs[3], data_metricas, "Métricas de Sucesso")

    # Ajustar layout e salvar PDF
    plt.tight_layout(pad=3.0)
    pdf_pages.savefig(fig, bbox_inches='tight')
    pdf_pages.close()

    print(f"PDF '{pdf_filename}' gerado com sucesso!")
    return pdf_filename

def visualizar_pdf(pdf_path):
    print("\nOpções de Visualização:")
    resposta = input("[A]brir PDF | [B]aixar PDF | [V]isualizar Resumo: ").upper()

    if resposta == 'A':
        webbrowser.open(pdf_path)
    elif resposta == 'B':
        print(f"\nArquivo disponível em:\n{os.path.abspath(pdf_path)}")
    elif resposta == 'V':
        print("\n" + "="*80)
        print("PLANO SCRUM - AEROVISION ANALYTICS PARA AEROPORTOS")
        print("-"*80)
        print("SPRINT 1 (2 semanas): Diagnóstico de Mercado")
        print("SPRINT 2 (3 semanas): Prototipagem de Solução")
        print("SPRINT 3 (4 semanas): Abordagem Comercial")
        print("\nEQUIPE:")
        print("- Product Owner: Head de Produto da AeroVision")
        print("- Scrum Master: Analista Líder de Dados")
        print("- Dev Team: Cientista de Dados, Engenheiro de Dados, Designer de UX, Representante Comercial")
        print("="*80)
    else:
        print("Opção inválida. O PDF será aberto automaticamente.")
        webbrowser.open(pdf_path)

if __name__ == "__main__":
    pdf_file = gerar_pdf_scrum()
    visualizar_pdf(pdf_file)
```
